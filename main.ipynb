#  Best Graphs to Compare `y_test` vs `y_pred` 

### 1. Confusion Matrix (Most Common)

A confusion matrix is a simple table that helps you understand how well a classification model performs by comparing the actual labels with the predicted ones. It shows four key values: True Positives (TP) where the model correctly predicted the positive class, True Negatives (TN) where it correctly predicted the negative class, False Positives (FP) where it wrongly predicted positive, and False Negatives (FN) where it wrongly predicted negative. These values give a more detailed picture of what your model is doing right or wrong.

Using the confusion matrix, you can calculate other important metrics like accuracy, precision, recall, and F1-score, which help you judge how good your model is  not just overall, but also in specific situations like when false positives or false negatives are more serious. This makes the confusion matrix a powerful tool for analyzing classification performance in a meaningful way.

```python
# Importing necessary libraries
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay  # For evaluating the model using a confusion matrix

# Step 1: Compute the confusion matrix
# This compares the actual labels (y_test) with the predicted labels (y_test_pred)
# It returns a 2x2 matrix in a binary classification problem:
# For example, if classes are ["Rock", "Mine"]:
# [
#   [True Negatives, False Positives],
#   [False Negatives, True Positives]
# ]
cm = confusion_matrix(y_test, y_test_pred)

# Step 2: Create a ConfusionMatrixDisplay object
# We give the confusion matrix and the names of each category (like "Rock" and "Mine")
# This helps us label the rows and columns so we know what each number in the matrix means.
# For example:
# - The top-left box will show how many times the model correctly predicted "Rock"
# - The bottom-right will show correct "Mine" predictions
# So, adding labels makes the confusion matrix easier to read and understand.
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Rock", "Mine"])

# Step 3: Plot the matrix
# `cmap='Reds' sets the color map to various shades of red; darker means higher numbers
disp.plot(cmap='Purples')

# Step 4: Add a title to the plot for better understanding
plt.title("Confusion Matrix")

# Step 5: Show the plot on the screen
plt.show()
```

### 2. ROC Curve (Good for seeing model performance over thresholds)

The ROC (Receiver Operating Characteristic) curve is a graph that helps us understand how well a classification model can separate two classes  like telling apart cats and dogs. It does this by plotting the True Positive Rate (how often the model correctly identifies the positive class) against the False Positive Rate (how often it mistakenly thinks a negative is positive) at different confidence thresholds. The model might not always be 100% confident, so the ROC curve shows how its performance changes as we adjust the threshold for making decisions.

The AUC (Area Under the Curve) is a number between 0 and 1 that summarizes the ROC curve into a single value. AUC close to 1 means the model is very good at distinguishing between the two classes, while 0.5 means it's no better than random guessing. The higher the AUC, the better the model's overall performance across all decision thresholds. This makes the ROC curve and AUC useful tools for comparing and evaluating different models.

```python
# Import necessary functions for ROC curve and AUC computation
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt  # Needed for plotting

# Get the predicted probabilities for the positive class (class 1)
# In machine learning, classes are the labels your model predicts  like 0 = not spam and 1 = spam, where class 1 is usually the "positive" outcome you're most interested in. 
# model.predict_proba(X_test) gives probabilities for both classes; we take the second column (index 1), which helps us evaluate how confidently the model predicts the positive class  especially when plotting the ROC curve.
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:, 1])

# Calculate the Area Under the Curve (AUC) from the FPR and TPR
roc_auc = auc(fpr, tpr)

# Plot the ROC curve using FPR and TPR
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")

# Plot a diagonal dashed line representing random performance (AUC = 0.5)
plt.plot([0, 1], [1, 0], linestyle='--', color='gray')

# Add title and labels to the plot
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

# Show the legend (which includes the AUC value)
plt.legend()

# Show grid lines for better readability
plt.grid()

# Display the final ROC plot
plt.show()
```

### 3. Precision-Recall Curve (for imbalanced data)

The Precision-Recall (PR) curve is a useful tool to evaluate a model's performance, especially when dealing with imbalanced datasets. It helps you understand how well the model identifies the positive class by showing the trade-off between precision and recall at different classification thresholds. This curve is particularly important when the costs of false positives and false negatives are not equal, or when you want to focus specifically on the positive class.

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

precision, recall, _ = precision_recall_curve(y_test, model.predict_proba(X_test)[:, 1]) # The _ is used as a placeholder for a value that is returned by the precision_recall_curve() function but is not needed or not used.
ap = average_precision_score(y_test, model.predict_proba(X_test)[:, 1])

plt.plot(recall, precision, label=f"AP = {ap:.2f}")
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.grid()
plt.show()
```

### 4. Bar Plot of Actual vs Predicted Count

A bar plot of Actual vs. Predicted Counts is a visual representation that compares the frequency of actual and predicted values in a classification problem. It helps you see how well your model's predictions align with the true labels.

```python
import pandas as pd
import seaborn as sns

# Create a DataFrame to compare actual vs predicted values
# 'Actual' column contains the true labels (y_test), and 'Predicted' contains the model's predicted labels (y_test_pred)
compare_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

# The `melt()` function reshapes the DataFrame from wide format (separate 'Actual' and 'Predicted' columns) to long format
# In the melted DataFrame, 'value' will represent the actual or predicted values, 
# The x='value' specifies that the x-axis of the plot should represent the values in the value column of the melted DataFrame.
# and 'variable' will indicate whether the value came from the 'Actual' or 'Predicted' column
sns.countplot(data=compare_df.melt(), x='value', hue='variable')

# Add title and axis labels for the plot
plt.title("Actual vs Predicted Class Counts")  # Title of the plot
plt.xlabel("Class")  # Label for the x-axis (representing the class)
plt.ylabel("Count")  # Label for the y-axis (representing the count of occurrences)

# Remove the legend title (since it's not necessary for clarity)
plt.legend(title="")

# Show the plot
plt.show()
```
